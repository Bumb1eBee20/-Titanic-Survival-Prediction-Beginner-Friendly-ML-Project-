```python
titanic.isnull().sum()
plt.figure(figsize=(12,6))
sns.heatmap(titanic.isnull(),cbar=False,cmap="Blues")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
titanic["Age"]=titanic["Age"].fillna(titanic["Age"].mean())
```

```python
titanic["Embarked"]=titanic["Embarked"].fillna(titanic["Embarked"].mode()[0])
```

```python
titanic.isnull().sum()
```

```python
titanic=titanic.drop(columns="Cabin")
```

```python
titanic.isnull().sum()
```

```python
plt.figure(figsize=(5,2))
sns.countplot(data=titanic,x="Sex",palette="Set1")
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(5,2))
sns.countplot(data=titanic,x="Pclass",palette = "Set2")
plt.title("Passenger Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(5,2))
sns.countplot(data=titanic,x="Embarked",palette="Set3")
plt.title("Port of Embarked Distribution")
plt.show()
```

```python
titanic.duplicated().sum()
```

```python
titanic["Survived"]=titanic["Survived"].map({0:"Died",1:"Survived"})
```

```python
#survival count
sns.countplot(data=titanic,x="Survived",palette="Blues")
plt.title("Survival Count")
plt.show()
```

```python
#2. Survival by Gender
sns.countplot(data=titanic,x="Sex",hue="Survived",palette="Blues")
plt.title("Survival by Gender")
plt.legend(labels=["Died","Survived"])
plt.show()
```

```python
#3. Survival by Passenger Class
sns.countplot(data=titanic,x="Pclass",hue="Survived",palette="Set1")
```

```python
#4. Survival by Age
plt.figure(figsize=(50,20))
sns.countplot(data=titanic,x='Age',hue="Survived",palette="Set2")
plt.show()
```

```python
#5. plt.figure(figsize=(10, 8))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
